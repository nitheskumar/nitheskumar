       2.1 CUSTOMER SEGMENTATION :

```python
df.head(3)
```

          2.1.1 TYPES OF CUSTOMERS AND NUMBERS OF CUSTOMERS :

```python
# types of customers
Types_of_customers = df['Segment'].unique()
print(Types_of_customers)
```

```python
# numbers of customers
numbers_of_customers= df['Segment'].value_counts().reset_index()
print(numbers_of_customers)
```

2.2 PLOTTING A PIE CHART :

```python
#plotting a pie chart

plt.pie(numbers_of_customers['count'], labels=numbers_of_customers['Segment'], autopct='%1.1f%%')

#set pie chart lables
plt.title('Distribution of Customers')
#generate a pie chart
plt.show()
```

2.3 SALES PER CUSTOMER SEGMENT ANALYSIS :

```python
sales_per_category= df.groupby('Segment')['Sales'].sum().reset_index()
print(sales_per_category)

sales_per_category= sales_per_category.rename(columns={'Segment': 'Customers Type','sales':'Total Sales'})
print(sales_per_category)
```

```python
sales_per_category= df.groupby('City')['Sales'].sum().reset_index()
print(sales_per_category)

sales_per_category= sales_per_category.rename(columns={'City': 'City name','sales':'Total Sales'})
print(sales_per_category)
```

2.2 PLOTTING A BAR CHART :

```python
#plotting a bar chart
plt.bar(sales_per_category['Customers Type'], sales_per_category['Sales'])

#set bar chart lables
plt.title('Sales per customers category')
plt.xlabel('customer type')
plt.ylabel('Total sales')

#generate a pie chart
plt.show()
```

3 CUSTOMER LOYALTY ANALYSIS :

```python
df.head()
```

```python
# Group data according to: Customer ID,customer Name,segment and calculate freq. of the that orders
customer_orders_freq=df.groupby(['Customer ID','Customer Name','Segment',]) ['Order ID'].count().reset_index()

#Rename the order ID columns
customer_orders_freq.rename(columns={'Order ID' : 'Total orders'}, inplace=True)

#Identify repeat customers
repeat_customers = customer_orders_freq[customer_orders_freq['Total orders']>= 1]


#sort repeat customer in descenting order
sorted_repeat_customers= repeat_customers.sort_values(by='Total orders', ascending=False)

#print statement
print(sorted_repeat_customers.head(10).reset_index(drop=True))
```

```python
# Group data according to: Customer ID,customer Name,segment and calculate freq. of the that orders
customer_sales=df.groupby(['Customer ID','Customer Name',])['Sales'].sum().reset_index()

#sort repeat customer in descenting order
Top_spenders = customer_sales.sort_values(by='Sales', ascending=False)

#print statement
print(customer_sales.head(10).reset_index(drop=True))
```

```python
#Group data by product sub-category vs sales
pdt_subcategory = df.groupby(['Sub-Category'])['Sales'].sum().reset_index()

#sotring in descending order
top_pdt_subcategory = pdt_subcategory.sort_values(by='Sales',ascending=False)

print(top_pdt_subcategory)
```

```python
top_pdt_subcategory = top_pdt_subcategory.sort_values(by='Sales',ascending=True)

#plotting a bar chart
plt.barh(top_pdt_subcategory['Sub-Category'], top_pdt_subcategory['Sales'])

#set bar chart lables
plt.title('Top product sub-category based on sales')
plt.xlabel('product sub-categories')
plt.ylabel('Total Sales')

#generate a pie chart
plt.show()
```
