4.states per recover,active and Deceased :

```python
data.head(10)
```

```python
covid_per_states= data.groupby('State')['Confirmed'].sum().reset_index()
print(covid_per_states)

covid_per_states= covid_per_states.rename(columns={'state': 'Confirmed','confirmed':'Total confirmed'})
print(covid_per_states)
```

```python
Deceased_per_states= data.groupby('State')['Deceased'].sum().reset_index()
print(Deceased_per_states)

Deceased_per_states= Deceased_per_states.rename(columns={'state': 'Deceased','Deceased':'Total Deceased'})
print(Deceased_per_states)
```

```python
Recovered_per_states= data.groupby('State')['Recovered'].sum().reset_index()
print(Recovered_per_states)

Recovered_per_states= Recovered_per_states.rename(columns={'state': 'Recovered','Recovered':'Total Recovered'})
print(Recovered_per_states)
```

```python
Active_per_states= data.groupby('State')['Active'].sum().reset_index()
print(Active_per_states)

Active_per_states= Active_per_states.rename(columns={'state': 'Active','Active':'Total Active'})
print(Active_per_states)
```

```python
Recovered_per_District= data.groupby('District')['Recovered'].sum().reset_index()
print(Recovered_per_District)

Recovered_per_District= Recovered_per_District.rename(columns={'District': 'Recovered','Recovered':'Total Recovered'})
print(Recovered_per_District)
```
